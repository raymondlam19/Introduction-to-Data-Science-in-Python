```python
import pandas as pd

# rows
customers = [('Alan', 'Dog food', 1.0, None),
             ('Betty', 'Cat food', 2.0, None),
             ('Chris', 'Bird food', 3.0, None),
             ('David', 'Human food', 4.0, 'UK'),
             ('Evid', 'Wolf food', 5.0, 'USA')]

# Table
df = pd.DataFrame(customers,
                 columns=['Name', 'Item purchased', 'Cost', 'Location'],
                 index=['Store1', 'Store1', 'Store3', 'Store4', 'Store5'])
df
```

```python
costs = df['Cost']
costs
```

```python
costs += 10
costs
```

```python
import pandas as pd
df = pd.read_csv('olympics.csv')
df.head()
```

```python
# choose which column to be index, which row to be header
df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
df.head()
```

```python
df.columns
```

```python
for col in df.columns:
    print(col)
```

```python
for col in df.columns:
    print(col[:2])
```

```python
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2] == '02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2] == '03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1] == '№':
        df.rename(columns={col:col[2:]}, inplace=True)
df.head()
```

# Querying a DataFrame

```python
df['Gold'] > 0
```

```python
only_gold = df.where(df['Gold'] > 0)
only_gold.head()
```

```python
# NaN will be ignored
only_gold['Gold'].count()
```

```python
df['Gold'].count()
```

```python
# drop NaN rows
only_gold = only_gold.dropna()
only_gold.head()
```
